```python
import numpy as np

# Step function (activation)
def step_function(x):
    return 1 if x >= 0 else 0

# Training data for digits 0–9 (ASCII binary representation)
def get_ascii_binary(c):
    return [int(b) for b in format(ord(c), '08b')]

training_data = [
    {'input': get_ascii_binary('0'), 'label': 1},  # even
    {'input': get_ascii_binary('1'), 'label': 0},  # odd
    {'input': get_ascii_binary('2'), 'label': 1},
    {'input': get_ascii_binary('3'), 'label': 0},
    {'input': get_ascii_binary('4'), 'label': 1},
    {'input': get_ascii_binary('5'), 'label': 0},
    {'input': get_ascii_binary('6'), 'label': 1},
    {'input': get_ascii_binary('7'), 'label': 0},
    {'input': get_ascii_binary('8'), 'label': 1},
    {'input': get_ascii_binary('9'), 'label': 0},
]
```

```python
# Initialize weights and bias
weights = np.zeros(8)
bias = 0
learning_rate = 1

# Training the perceptron
for epoch in range(10):  # train for multiple epochs
    for data in training_data:
        x = np.array(data['input'])
        y = data['label']
        z = np.dot(weights, x) + bias
        pred = step_function(z)
        error = y - pred
        weights += learning_rate * error * x
        bias += learning_rate * error
```

```python
# Prediction phase
digit = input("Enter a digit (0–9): ")

if digit not in '0123456789':
    print("Invalid input.")
else:
    x_test = np.array(get_ascii_binary(digit))
    z = np.dot(weights, x_test) + bias
    result = step_function(z)
    label = "Even" if result == 1 else "Odd"
    print(f"The digit {digit} is {label}.")
```
