```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-5,5)
y = sigmoid(x) 
plt.title('Sigmoid function')
plt.plot(x,y,'r')
plt.show()
```

```python
def TanH(x):
    return np.tanh(x)

x = np.linspace(-5,5)
y = TanH(x) 
plt.title('Tanh function')
plt.plot(x,y,'r')
plt.show()
```

```python
def relu(x):
    return np.maximum(0,x)
x = np.linspace(5,-5)
y = relu(x)
plt.plot(x,y,'r')
plt.title('ReLU Function')
plt.show()
```

```python
def softmax(x):
    exp_values = np.exp(x - np.max(x, axis=0))
    return exp_values / np.sum(exp_values, axis=0)

x = np.linspace(5,-5)
y = softmax(x)
plt.plot(x,y,'r')
plt.title('Softmax Function')
plt.show()
```
